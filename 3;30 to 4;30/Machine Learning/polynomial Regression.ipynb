```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = 6 * np.random.rand(200, 1) - 3
y = 0.8 * x ** 2 + 0.9 * x + 2 + np.random.randn(200, 1)
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
plt.scatter(X_train, y_train, label = "Train", color = "red")
plt.scatter(X_test, y_test, label = "Test", color = "blue")
plt.plot(X_test, model.predict(X_test), label = "Predicted", color = "red")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures( degree = 1 )
```

```python
X_train_new = poly.fit_transform(X_train)
```

```python
X_train_new
```

```python
X_test_new = poly.fit_transform(X_test)
```

```python
model_new = LinearRegression()
```

```python
model_new.fit(X_train_new, y_train)
```

```python
y_pred_new = model_new.predict(X_test_new)
```

```python
r2_score(y_test, y_pred_new)
```

```python
plt.scatter(X_train, y_train, label = "Train", color = "red")
plt.scatter(X_test, y_test, label = "Test", color = "blue")
plt.plot(X_test, model_new.predict(X_test_new), label = "Predicted", color = "red")
plt.legend()
plt.show()
```
