```python
import matplotlib.pyplot as plt
```

# hist Plot

```python
no=[36,58,53,16,29,75,87,40,50,65,26,11,32,16,69,46,82,70,21,85]
plt.hist(no,"auto",(0,100),histtype="bar",rwidth=0.8,label="Python", orientation = "horizontal")
plt.legend()
plt.show()

# step, stepfilled, bar, barstacked
```

# stem plot

```python
x = [1,2,3,4,5,6,7,8]
y = [10,20,30,40,50,60,70,80]
plt.stem(x, y, markerfmt = "*", orientation = "horizontal")
plt.show()
```

# Box Plot

```python
x = [10,20,30,40,50,60,70,80]
plt.boxplot(x, notch = True, vert = False)
plt.show()
```

# Stack Plot

```python
x = [1,2,3,4,5,6,7,8]
y = [10,20,30,10,50,20,70,50]
plt.stackplot(x, y)
plt.show()
```
