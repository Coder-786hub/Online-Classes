```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,50,60]
plt.plot(x, y)
plt.title("MATPLOT PLOT" )
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

# BAR Plot


```python
x = ["Python", "Java", "C", "C++", "HTML"]
y = [70, 45, 49, 30, 65]
z = ["red", "green", "blue", "red", "green"]
plt.bar(x, y, color = z, width = 0.4, label = x)
plt.xlabel("Programming Language", fontsize = 20)
plt.ylabel("Importante", fontsize = 20)
plt.title("Showing Programming Language", fontsize = 20)

plt.legend()
plt.show()
```
