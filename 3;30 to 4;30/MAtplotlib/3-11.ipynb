```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 5))
x = [1,2,3,4,5,6,7,8]
y = [10,20,30,40,50,60,70,80]
# color = ["red","green","blue","black","red","green","blue","black"]
size = [150, 200, 250, 300, 350, 400, 450, 500]
color = [20, 30, 50, 60, 70, 90, 80, 100]
plt.scatter(x,y, c = color, s = size, cmap = "Accent_r" )
plt.show()
```

# Pie Chart

```python
label = ["Rohit", "Virat", "Hardik", "Surya", "Kuldeep", "Jasprit", "Rishabh", "Ashwin"]
plt.pie(y, labels = label, autopct = "%0.1f%%")
# plt.legend()
plt.show()
```

```python
color = ["red","green","blue","pink","red","green","blue","pink"]
plt.hist(y, rwidth = 0.2, colors =color )
plt.show()
```
