```python
import pandas as pd
```

```python
data=pd.DataFrame({
    'outlook': ['sunny', 'overcast', 'rainy','sunny',"rainy"] * 10,  
    'humidity': ['high', 'normal'] * 25,  
    'wind': ['weak', 'strong'] * 25, 
    'play': ['yes', 'no'] * 25  
})
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
outlook=LabelEncoder()
humidity=LabelEncoder()
wind=LabelEncoder()
play=LabelEncoder()
```

```python
data["outlook"]=outlook.fit_transform(data["outlook"])
data["humidity"]=outlook.fit_transform(data["humidity"])
data["wind"]=outlook.fit_transform(data["wind"])
data["play"]=outlook.fit_transform(data["play"])
```

```python
data
```

```python
features_colm=["outlook","humidity","wind"]
x=data[features_colm]
y=data.play
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# lr = DecisionTreeClassifier(criterion="gini")
lr = DecisionTreeClassifier(criterion="entropy")
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)   # 0 no   1 yes  outlook[sunny 2, overcast 0, rainy 1]
```

```python
x_test
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(lr)
```
